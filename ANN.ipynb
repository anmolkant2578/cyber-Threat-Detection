```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import sklearn_relief as relief
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'cyber121.csv')
```

```python
df.head()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from mlxtend.evaluate import confusion_matrix
from sklearn.metrics import classification_report
```

```python
from sklearn.neural_network import MLPClassifier

# Create an ANN model (MLPClassifier)
ann = MLPClassifier(hidden_layer_sizes=(100,), max_iter=200, random_state=10, alpha=0.0001)

# Train the ANN model
ann.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Predict using the trained model
y_pred = ann.predict(X_test)  # assuming you have a test set (X_test)

# Calculate and print the accuracy score
accuracy = accuracy_score(y_test, y_pred)  # y_test is the true labels of the test set
print(f"Accuracy Score: {accuracy:.4f}")
```
